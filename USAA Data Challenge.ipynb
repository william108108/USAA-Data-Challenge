```python
def count_delivered(df):
    # initialize count
    count = 0 
    # create empty dictionary, key is the row number of the property, value is the column number
    # when that property is delivered (will use in next part)
    delivered_month = {}
    for i in range(len(df)):
        for j in range(list(df.columns).index("Apr-08"), len(list(df.columns))):
            if type(df.iloc[i,j]) == str:
                if df.iloc[i,j] == 'LU':
                    count += 1
                    delivered_month[i] = j
                    break
                elif df.iloc[i,j] == "UC/LU":
                    count += 1
                    delivered_month[i] = j
                    break
                else:
                    break
            elif type(df.iloc[i, j]) == float:
                continue
    return count, delivered_month
            
count_MSA1, delivered_month_MSA1 = count_delivered(MSA1_prop_stat)
count_MSA2, delivered_month_MSA2 = count_delivered(MSA2_prop_stat)
print(count_MSA1)
print(count_MSA2)
```

### The number of properties delivered is 270 for Austin-Round Rock, TX and 9 for Akron, OH since April 2008.

## Calculate average lease-up time

```python
MSA1_occ = pd.read_excel("MSA1.xlsx", sheet_name="Occ & Concession", header = 4)
MSA2_occ = pd.read_excel("MSA2.xlsx", sheet_name="Occ & Concession", header = 4)
```

```python
MSA1_occ.head()
```

```python
MSA2_occ.head()
```

```python
# Find where occupancy columns end, filter out other columns
list(MSA1_occ.columns).index("Apr-08.1")
```

```python
MSA1_occ = MSA1_occ.iloc[:,0:168]
MSA2_occ = MSA2_occ.iloc[:,0:168]
```

```python
def lease_up(df, delivered_month):
    # create empty list for each property's lease-up time
    month = []
    for k,v in delivered_month.items():
        for j in range(list(df.columns).index("Apr-08"), len(list(df.columns))):
            if df.iloc[k, j] >= 0.9: # find first month its occupancy rate reaches 90% or above
                month.append(j - v) # find difference between column number 
                break
            else:
                continue
    return np.mean(month)

avg_month_MSA1 = lease_up(MSA1_occ, delivered_month_MSA1)
avg_month_MSA2 = lease_up(MSA2_occ, delivered_month_MSA2)
print(avg_month_MSA1)
print(avg_month_MSA2)
```

### The average lease-up time is 12.72 months for Austin-Round Rock, TX and 8.75 months for Akron, OH.

```python
# create an empty dataframe, set columns, and append all states' dataframes to this dataframe
df = pd.DataFrame(columns =['State_in', 'County_in', 'State_from', 'County_from', 'State', 'County',
                                             'Number of Returns', 'Number of Exemptions', 'Adjusted Gross Income'])
for file in os.listdir():
    if file.endswith('xls'):
        state = pd.read_excel(file)
        state.columns = ['State_in', 'County_in', 'State_from', 'County_from', 'State', 'County',
                                             'Number of Returns', 'Number of Exemptions', 'Adjusted Gross Income']
        state = state.iloc[7:,:]
        df = df.append(state, ignore_index = True)
```

```python
df
```

```python
# found error entries, drop them
df = df[df['Number of Returns'] != 'd60']
df = df[df['Number of Returns'] != 'r 81']
```

```python
# convert specific column types to integers
df = df.astype({'State_in': 'int64', 'State_from': 'int64', 'Number of Returns': 'int64',
               'Number of Exemptions': 'int64', 'Adjusted Gross Income': 'int64'})
```

```python
# Filter out total migration rows by selecting State_from < 55, and non-migrants (when state_in = state_from and 
# County_in = County_from)

# Find maximum number of exemptions
np.max(df[(df['State_in'] != df['State_from']) & (df['County_in'] != df['County_from']) & 
   (df['State_from'] < 55)]['Number of Exemptions'])
```

## Task 2.4

```python
# All counties in NY metro area
NY_metro = [
'Bergen County',
'Dutchess County',
'Essex County',
'Fairfield County',
'Hudson County'
'Hunterdon County',
'Litchfield County',
'Mercer County',
'Middlesex County',
'Monmouth County',
'Morris County',
'Nassau County',
'New Haven County',
'Ocean County',
'Orange County',
'Passaic County',
'Pike County',
'Putnam County',
'Rockland County',
'Somerset County',
'Suffolk County',
'Sussex County',
'Ulster County',
'Union County',
'Westchester County']
```

```python
# Find total in-migrations into NY metro counties
total_in = 0
for county in NY_metro:
    county_column = county + ' Tot Mig-US & For'
    if df[df['County'] == county_column]['Number of Exemptions'].empty:
        total_in = total_in
    else:
        total_in += df[df['County'] == county_column]['Number of Exemptions'].iloc[0]
    
print(total_in)
```

```python
# Find total out-migrations from NY metro counties
total_out = 0
for county in NY_metro:
    if df[df['County'] == county].empty:
        total_out = total_out
    else:
        for i in range(len(df[df['County'] == county])):
            total_out += df[df['County'] == county]['Number of Exemptions'].iloc[i]
print(total_out)
```

```python
net_migration = total_in - total_out
print(net_migration)
```

### Net migration for NY metropolitan area is -576107.
